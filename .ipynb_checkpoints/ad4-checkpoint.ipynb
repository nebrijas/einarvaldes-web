```python
import pandas as pd
```

```python
url = "https://api.covid19api.com/countries"
```

```python
url
```

```python
df = pd.read_json(url)
```

```python
df
```

```python
df[df["Country"] == "Spain"]
```

```python
url_rt_es = "https://api.covid19api.com/country/spain/status/confirmed/live"
df_rt_es = pd.read_json(url_rt_es)
df_rt_es
```

```python
df_rt_es.head()
```

```python
df_rt_es.tail()
```

```python
df_rt_es.set_index('Date')['Cases'].plot(title="Casos de Covid-19 en España desde 20/1/2020 hasta 29/6/2022",kind = "area")
```
